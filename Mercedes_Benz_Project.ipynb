##1. Removing variables whose variance is equal to zero.

```python
# Identify columns with zero variance in df_train
zero_var_cols = df_train.columns[df_train.nunique() == 1]

# Drop those columns
df_train = df_train.drop(columns=zero_var_cols)

# Print the removed columns
print(f"Removed {len(zero_var_cols)} zero-variance columns:")
print(zero_var_cols.tolist())
```

```python
# Drop extra columns from test data to match train data
df_test = df_test.drop(columns=['X347', 'X297', 'X293', 'X107', 'X290', 'X93', 'X330', 'X11', 'X289', 'X235', 'X233', 'X268'])

# Verify that test data now has the same columns as train data
print("Test dataset columns after dropping extras:", df_test.columns)
```

##3. Applying Label Encoder in Train & Test dataset.

```python
# Identify categorical columns
categorical_cols = df_train.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols.tolist())
```

```python
label_encoders = {}  # Dictionary to store encoders

for col in categorical_cols:
    le = LabelEncoder()
    df_train[col] = le.fit_transform(df_train[col])  # Fit & transform training data
    df_test[col] = df_test[col].map(lambda x: le.transform([x])[0] if x in le.classes_ else -1)  # Handle unseen categories
    label_encoders[col] = le  # Store encoder for future use
```

## 4. Performing dimensionality reduction

```python
# Exclude target column ('y') and apply standard scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(df_train.drop(columns=['y']))
X_test_scaled = scaler.transform(df_test)

# Confirm the data is scaled properly
print("Shape of X_train_scaled:", X_train_scaled.shape)
print("Shape of X_test_scaled:", X_test_scaled.shape)
```

```python
# Define the number of principal components to keep
n_components = 12
pca = PCA(n_components=n_components)

# Fit PCA on training data and transform both train & test sets
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Print the new shape after PCA
print(f"Reduced train data shape: {X_train_pca.shape}")
print(f"Reduced test data shape: {X_test_pca.shape}")
```

## 5.Predict your test_df values using XGBoost.

```python
from xgboost import XGBRegressor

# Initialize the model
model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)

# Train the model using PCA-transformed features
model.fit(X_train_pca, df_train['y'])  # Target column is 'y'
```

```python
y_pred = model.predict(X_test_pca)

# Display first few predictions
print("Predicted values for test_df:", y_pred[:10])
```
